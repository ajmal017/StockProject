```python
def Pricing(ticker, dstart):

    df = yf.download(ticker, 
                          start=dstart, 
                          
                          progress=True, group_by = 'Ticker')

    df = df.reset_index()
    df2 = df.melt(id_vars = 'Date', var_name = ['Ticker','Type'])
    df2 = df2.set_index(['Date','Ticker'])
    df2 = df2.pivot(columns = 'Type').reset_index()
    df2.columns = df2.columns.droplevel()
    columnlist = list(df2.columns)
    columnlist[0] = 'Date'
    columnlist[1] = 'Ticker'

    df2.columns = columnlist
    
    
    return df2
```

```python
dfPrice = Pricing(list(df.Symbol), '2020-07-01')
```

```python
dfPrice = dfPrice.sort_values(by = ['Ticker', 'Date'])
```
